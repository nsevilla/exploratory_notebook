##### Plot 2d histos, per band, superimpose a scatter plot with statistical info 

```python
from matplotlib.colors import LogNorm
nbins2d=(100,100)
minmag = 16
maxmag = 25
fig,axes = subplots(1, 5, figsize=(15,3))
for counter,band in enumerate(['G','R','I','Z','Y']):
    residuals = np.empty([])
    residuals_error = np.empty([])
    magbin = np.empty([])
    #fig = plt.figure()
    xstring1 = 'MAG_AUTO_' + band
    xstring2 = 'XCORR_SFD98_' + band
    ystring1 = 'MAG_MODEL_' + band 
    ystring2 = 'MAG_AUTO_' + band 
    for maglo in xrange(minmag, maxmag):
        maghi = maglo + 1
        mask_hi = tdata['MAG_AUTO_' + band] < maghi
        mask_lo = tdata['MAG_AUTO_' + band] > maglo
        mask_99 = tdata[ystring1] != 99
        mask_gal = abs(tdata['WAVG_SPREAD_MODEL_I']) > 0.003
        mask = mask_hi & mask_lo & mask_flags & mask_sm1 & mask_detmodel_flags & mask_99 & mask_gal
        magbin = np.append(magbin,float(maghi+maglo)/2) 
        residuals = np.append(residuals,mean(tdata[ystring1][mask]-tdata[ystring2][mask]))
        residuals_error = np.append(residuals_error,
                                    std(tdata[ystring1][mask]-tdata[ystring2][mask]))
    axes[counter].hist2d(tdata[xstring1][mask_global]-tdata[xstring2][mask_global],
                         tdata[ystring1][mask_global]-tdata[ystring2][mask_global],
                         nbins2d,range=[[15,29],[-8,8]],cmap='jet', norm=LogNorm(), label=band)
    axes[counter].errorbar(magbin[1:],residuals[1:],yerr=residuals_error[1:],c='white',
                           linestyle="None",fmt='o')
    axes[counter].set_xlabel("Dereddened MAG_AUTO_" + band)
    axes[counter].set_ylabel("MAG_MODEL - MAG_AUTO in " + band + ' band')
#filename = "mag_model_auto_residuals_vs_mag_auto_gals.png'
#fig.savefig(filename)
```
