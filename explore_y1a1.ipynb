#Exploring Y1A1 dataset

This IPython (Jupyter) notebook provides examples on how to plot some basic figure from the downloaded FITS file. 


#Import modules

```python
import numpy as np
from astropy.io import fits
import matplotlib
import matplotlib.pyplot as plt
```

#Open file

```python
hdulist = fits.open('exploratory_y1a1_1.fits',memmap=True)
```

```python
print hdulist[1].columns
```

#Load data into table

```python
tdata = hdulist[1].data
```

#Define global selection masks

```python
mask_hi = tdata['MAG_AUTO_I'] < 30
mask_lo = tdata['MAG_AUTO_I'] > 15
mask_flags = tdata['FLAGS_I'] < 4
mask_detmodel_flags = tdata['FLAGS_DETMODEL_I'] == 0
mask_sm1 = tdata['SPREAD_MODEL_I'] != 1
#mask_nit = tdata['NITER_MODEL_I'] > 0 
mask_nep_eq0 = tdata['NEPOCHS_I'] == 0
mask_nep_gt0 = tdata['NEPOCHS_I'] > 0
mask_gal = tdata['WAVG_SPREAD_MODEL_I'] > 0.003
mask_sta = abs(tdata['WAVG_SPREAD_MODEL_I']) < 0.003
mask1 = mask_hi & mask_lo & mask_flags & mask_sm1 & mask_nep_eq0
mask2 = mask_hi & mask_lo & mask_flags & mask_sm1 & mask_nep_gt0
mask_global = mask_hi & mask_lo & mask_flags & mask_sm1& mask_detmodel_flags & mask_gal
```

#Analysis and plots

##### Superimpose two histograms

```python
nbins=100
h1 = plt.hist(tdata['WAVG_SPREAD_MODEL_I'][mask1], nbins, alpha=0.5, 
              range=[-0.02,0.05],log=True,label='n_epochs_i==0')
h2 = plt.hist(tdata['WAVG_SPREAD_MODEL_I'][mask2], nbins, alpha=0.5, 
              range=[-0.02,0.05],log=True,label='n_epochs_i>0')
plt.xlabel('WAVG_SPREAD_MODEL_I')
plt.legend(loc='upper right')
```

##### Save a csv file with a selection

```python
nbins=30
range=[0,5]
#fig,axes = subplots(1, 2, figsize=(18,3))
mask_selection_1 = tdata['MAG_AUTO_I'] < 20
mask_selection_2 = tdata['MAG_AUTO_I'] > 18
mask = mask_global & mask_selection_1 & mask_selection_2
compstring = 'NEPOCHS_I'
#h0 = axes[0].hist(tdata[compstring][mask], nbins, range=range)
plt.hist(tdata[compstring][mask], nbins, range=range, alpha=0.5, label='No cuts', log=True)
mask_selection_3 = tdata['MAG_MODEL_I']-tdata['MAG_AUTO_I'] < -1.5
mask = mask & mask_selection_3
#print tdata['SPREAD_MODEL_I'][mask].shape[0]
#print tdata['COADD_OBJECTS_ID'][mask][20:30]
#print tdata['RA'][mask][20:30]
#print tdata['DEC'][mask][20:30]
a = np.asarray([tdata['COADD_OBJECTS_ID'][mask][1:100],tdata['RA'][mask][1:100],
                tdata['DEC'][mask][1:100]])
np.savetxt("outliers.csv", np.transpose(a), delimiter=",", fmt=['%10d','%10.6f','%10.6f']) 
### This selection can be uploaded and visualized in the Science Portal
#h1 = axes[1].hist(tdata[compstring][mask], nbins, range=range)
plt.hist(tdata[compstring][mask], nbins, range=range, alpha=0.5, label='Selection', log=True)
plt.legend(loc='upper right')
plt.xlabel(compstring)
```

##### Plot 2d histos, per band, superimpose a scatter plot with statistical info 

```python
from matplotlib.colors import LogNorm
nbins2d=(100,100)
minmag = 16
maxmag = 25
fig,axes = subplots(1, 5, figsize=(15,3))
for counter,band in enumerate(['G','R','I','Z','Y']):
    residuals = np.empty([])
    residuals_error = np.empty([])
    magbin = np.empty([])
    #fig = plt.figure()
    xstring1 = 'MAG_AUTO_' + band
    xstring2 = 'XCORR_SFD98_' + band
    ystring1 = 'MAG_MODEL_' + band 
    ystring2 = 'MAG_AUTO_' + band 
    for maglo in xrange(minmag, maxmag):
        maghi = maglo + 1
        mask_hi = tdata['MAG_AUTO_' + band] < maghi
        mask_lo = tdata['MAG_AUTO_' + band] > maglo
        mask_99 = tdata[ystring1] != 99
        mask_gal = abs(tdata['WAVG_SPREAD_MODEL_I']) > 0.003
        mask = mask_hi & mask_lo & mask_flags & mask_sm1 & mask_detmodel_flags & mask_99 & mask_gal
        magbin = np.append(magbin,float(maghi+maglo)/2) 
        residuals = np.append(residuals,mean(tdata[ystring1][mask]-tdata[ystring2][mask]))
        residuals_error = np.append(residuals_error,
                                    std(tdata[ystring1][mask]-tdata[ystring2][mask]))
    axes[counter].hist2d(tdata[xstring1][mask_global]-tdata[xstring2][mask_global],
                         tdata[ystring1][mask_global]-tdata[ystring2][mask_global],
                         nbins2d,range=[[15,29],[-8,8]],cmap='jet', norm=LogNorm(), label=band)
    axes[counter].errorbar(magbin[1:],residuals[1:],yerr=residuals_error[1:],c='white',
                           linestyle="None",fmt='o')
    axes[counter].set_xlabel("Dereddened MAG_AUTO_" + band)
    axes[counter].set_ylabel("MAG_MODEL - MAG_AUTO in " + band + ' band')
#filename = "mag_model_auto_residuals_vs_mag_auto_gals.png'
#fig.savefig(filename)
```
